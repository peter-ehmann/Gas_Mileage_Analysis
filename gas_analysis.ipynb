```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import pandas as pd
```

```python
link = "https://raw.githubusercontent.com/peter-ehmann/Gas_Mileage_Analysis/master/gas_log.csv"
local = "~/Documents/Programming/Gas/gas_log.csv"
data = pd.read_csv(link,
                   parse_dates=["date"],
                   index_col=["date"])
```

```python
data["mpg"] = data["miles"]/data["gallons"]
```

```python
data
```

```python
df = data.dropna(subset=["miles"])

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(df.index.values,
        df["mpg"], "-o",
        color="purple")

ax.set(#xlabel="Date",
       #ylabel="Miles Per Gallon",
       title="Miles Per Gallon over Time")

ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=5))
ax.xaxis.set_major_formatter(DateFormatter("%m/%d"))

plt.savefig("./figures/mpg_time.png")

plt.show()
```
