```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import pandas as pd
```

```python
data = pd.read_csv("/Users/pje/Documents/Programming/Gas/gas_log.csv",
                   parse_dates=["date"],
                   index_col=["date"])
```

```python
data["mpg"] = data["miles"]/data["gallons"]
```

```python
data
```

```python
df = data.dropna(subset=["miles"])

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(df.index.values,
        df["mpg"], "-o",
        color="purple")

ax.set(#xlabel="Date",
       #ylabel="Miles Per Gallon",
       title="Miles Per Gallon over Time")

ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=5))
ax.xaxis.set_major_formatter(DateFormatter("%m/%d"))

plt.show()
```
